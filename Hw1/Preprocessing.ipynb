# Basic Process

```python
from scipy.io import arff
import pandas as pd
```

```python
data = arff.loadarff('4year.arff')
df = pd.DataFrame(data[0])
```

```python
df['bankruptcy'] = (df['class']==b'1')
```

```python
df.head()
```

```python
df.describe()
```

```python
sum(df.bankruptcy == True)
```

# Create New DataFream and Set Labels

```python
df[['classb0','classb1']]=pd.get_dummies(df[['class']])
```

```python
df.head()
```

```python
df1 = df[['bankruptcy','Attr1','Attr2','Attr7','Attr10','classb0']]
```

```python
df1.head()
```

```python
df1.columns = ['bankruptcy','X1','X2','X7','X10','classb0']
```

```python
bankruptcy_mapping = {True:1,False:0}
df1['bankruptcy']=df['bankruptcy'].map(bankruptcy_mapping)
df1.head()
```

# Fill-in na

```python
from sklearn.preprocessing import Imputer

imr = Imputer(strategy = 'mean',axis=0)
imr = imr.fit(df1.values)
imputed_data = imr.transform(df1.values)
df2 = pd.DataFrame(imputed_data)
df2.columns = ['bankruptcy','X1','X2','X7','X10','classb0']
print(df2.head())
df2.isnull().sum()
```
